```python
from selenium import webdriver
from selenium.webdriver import ActionChains
import time
import re
import csv

# Windows users need to specify the path to chrome driver you just downloaded.
# You need to unzip the zipfile first and move the .exe file to any folder you want.
# driver = webdriver.Chrome(r'path\to\the\chromedriver.exe')
driver = webdriver.Chrome(r'C:\Users\fezro\chromedriver.exe')
# driver = webdriver.PhantomJS()
# Go to the page that we want to scrape
csv_file = open('whiskey.csv', 'w', encoding='utf-8', newline='')
writer = csv.writer(csv_file)
def loadMore():
    g=0
    while True:
        g+=1
        review_button = driver.find_element_by_css_selector('#loadMore')
        ActionChains(driver).move_to_element(review_button ).click(review_button ).perform()
        #ActionChains(driver).move_to_element(driver.find_element_by_id('#loadMore')).perform()
        #driver.execute_script("arguments[0].click();", review_button)
        
        #review_button.click()
        #review_button.click()
        print('click')
        if g>=4: break
        time.sleep(3)
        print('sleep')
for j in range(1,12):
    driver.get(
        "http://whiskyadvocate.com/ratings-reviews/?search=&submit=+&brand_id=0&rating=0&price=0&category="+str(j))
    #driver.get('http://whiskyadvocate.com/ratings-reviews/?search=&submit=+&brand_id=0&rating=0&price=0&category=1')
# html > body > div > div > div
    review_count = driver.find_element_by_css_selector('#main > .wrap').text
    #print(review_count)
# review_count = driver.find_element_by_xpath('html/body/div/div/div/main/div[@class="wrap"]').text
    review_count = int(re.search(r'[0-9]+',review_count).group(0))
    #print(review_count)
    if review_count % 6: click_count = review_count//6
    else: click_count = review_count//6-1


# Find all the reviews. The find_elements function will return a list of selenium select elements.
# Check the documentation here: http://selenium-python.readthedocs.io/locating-elements.html
    
    reviews = driver.find_elements_by_css_selector('#post-25')
    i = 0
    loadMore()
    
    # Iterate through the list and find the details of each review.
    for review in reviews:

            # Initialize an empty dictionary for each review
        review_dict = {}
			# Use try and except to skip the review elements that are empty.
			# Use relative xpath to locate the title.
			# Once you locate the element, you can use 'element.text' to return its string.
			# To get the attribute instead of the text of each element, use 'element.get_attribute()'
        try:

            title = review.find_element_by_css_selector('h1').text
            text = review.find_element_by_css_selector('.review-text').text
            rating = review.find_element_by_css_selector('h2').text
            author = review.find_element_by_css_selector(
			    '.review-text:nth-of-type(2)').text
            category = review.find_element_by_css_selector('p').text
        except:
            print('code 1')
            continue
        
        try:
            
            review_dict['Title'] = title.split(',')[0]
            review_dict['Text ']= text  
            review_dict['Rating']=rating         
            review_dict['Category']=category.split('|')[0]
            review_dict['Author'] = author.split(':')[1].split('(')[0]
            review_dict['Price ']=category.split('|')[1]
            review_dict['Season ']=author.split(':')[1].split('(')[1].split(')')[0]
              
        except:
            print("Missed a review item at category "+str(j)+", item "+str(i))
            continue
        try:
            review_dict['Alcohol ']=re.search('\d+(\.\d+)?%',title).group(0)
        except:
            print('No alcohol percentage')
    
        writer.writerow(review_dict.values())
        print('write')
        i+=1
        if i%24==0:loadMore()
```
